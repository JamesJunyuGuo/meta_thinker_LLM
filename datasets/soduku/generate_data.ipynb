```python
import random
import copy

SIZE = 4
SUBGRID = 2  # For 4x4: 2x2 subgrids
NUMBERS = list(range(1, SIZE + 1))

def is_valid(board, row, col, num):
    # Row and column
    if num in board[row]: return False
    if num in [board[r][col] for r in range(SIZE)]: return False
    # Subgrid
    start_row, start_col = row // SUBGRID * SUBGRID, col // SUBGRID * SUBGRID
    for r in range(start_row, start_row + SUBGRID):
        for c in range(start_col, start_col + SUBGRID):
            if board[r][c] == num: return False
    return True

def solve(board):
    for row in range(SIZE):
        for col in range(SIZE):
            if board[row][col] == 0:
                for num in NUMBERS:
                    if is_valid(board, row, col, num):
                        board[row][col] = num
                        if solve(board): return True
                        board[row][col] = 0
                return False
    return True

def generate_full_solution():
    board = [[0 for _ in range(SIZE)] for _ in range(SIZE)]
    nums = NUMBERS[:]
    for row in range(SIZE):
        for col in range(SIZE):
            random.shuffle(nums)
            for num in nums:
                if is_valid(board, row, col, num):
                    board[row][col] = num
                    if solve(copy.deepcopy(board)):
                        break
                    else:
                        board[row][col] = 0
            if board[row][col] == 0:
                return generate_full_solution()  # restart if stuck
    return board

def remove_numbers(board, holes=6):
    puzzle = copy.deepcopy(board)
    count = 0
    while count < holes:
        r, c = random.randint(0, SIZE - 1), random.randint(0, SIZE - 1)
        if puzzle[r][c] != 0:
            puzzle[r][c] = 0
            count += 1
    return puzzle

# === Generate Puzzle ===
solution = generate_full_solution()
puzzle = remove_numbers(solution, holes=6)

# === Print Puzzle and Solution ===
print("Puzzle:")
for row in puzzle: print(row)

print("\nSolution:")
for row in solution: print(row)
```

```python
import itertools
import random

def find_zero_sum_expressions(numbers):
    """Find all +/− sign assignments that sum to zero."""
    solutions = []
    for signs in itertools.product([-1, 1], repeat=len(numbers)):
        expression = [sign * num for sign, num in zip(signs, numbers)]
        if sum(expression) == 0:
            solutions.append(signs)
    return solutions

def format_expression(numbers, signs):
    """Format signs + numbers into a readable equation string."""
    parts = []
    for sign, num in zip(signs, numbers):
        parts.append(f"{'+' if sign > 0 else '-'}{num}")
    return " ".join(parts) + " = 0"

# Example usage
numbers = [1, 2, 3, 4, 6]
solutions = find_zero_sum_expressions(numbers)

print("Input numbers:", numbers)
print("Solutions found:", len(solutions))

for signs in solutions:
    print(format_expression(numbers, signs))
```
