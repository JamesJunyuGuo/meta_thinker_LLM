```python
import itertools

def solve_24(nums):
    # All possible operator combinations
    ops = ['+', '-', '*', '/']

    # Generate all permutations of numbers
    for num_perm in itertools.permutations(nums):
        # Generate all combinations of three operators
        for op_comb in itertools.product(ops, repeat=3):
            a, b, c, d = map(str, num_perm)
            op1, op2, op3 = op_comb

            # All possible parenthesis arrangements (5 in total)
            expressions = [
                f"(({a}{op1}{b}){op2}{c}){op3}{d}",
                f"({a}{op1}({b}{op2}{c})){op3}{d}",
                f"{a}{op1}(({b}{op2}{c}){op3}{d})",
                f"{a}{op1}({b}{op2}({c}{op3}{d}))",
                f"({a}{op1}{b}){op2}({c}{op3}{d})"
            ]

            for expr in expressions:
                try:
                    if abs(eval(expr) - 24) < 1e-6:
                        return True, expr
                except ZeroDivisionError:
                    continue
    return False, None

# Example
nums = [2,4,10,10]
found, expression = solve_24(nums)

if found:
    print("Found solution:", expression)
else:
    print("No solution found.")
```
